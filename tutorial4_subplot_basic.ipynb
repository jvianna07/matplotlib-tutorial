```python
n=201
x=np.linspace(0,4.0*np.pi,n)
y=np.sin(x)
y1=np.cos(x)
y2=np.tanh(x)
```

```python

ax1=plt.subplot(311)
ax2=plt.subplot(312)
ax3=plt.subplot(313)
ax1.plot(x,y)
ax2.plot(x,y1)
ax3.plot(x,y2)
plt.show()
```

```python
fig, ax =plt.subplots(3,1)
ax[0].plot(x,y)
ax[1].plot(x,y1)
ax[2].plot(x,y2)
plt.show()
```
