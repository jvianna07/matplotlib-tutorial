```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0, 2.0*np.pi, 101)
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.axis([0, 6.5, -1.1, 1.1])
plt.show()
```

```python
x = np.linspace(-5, 5, 400)
y = x**2

fig, ax = plt.subplots()

# Plota a função
ax.plot(x, y, label='$f(x) = x^2$')

# Define os limites dos eixos
ax.set_xlim(-6, 6)
ax.set_ylim(-5, 5)

# Remove os eixos tradicionais
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Coloca setas nas extremidades dos eixos
arrow_fmt = dict(facecolor='black', width=1, headwidth=8)

ax.annotate('', xy=(11, 0), xytext=(0, 0), arrowprops=arrow_fmt)  # seta eixo x positiva
ax.annotate('', xy=(0, 110), xytext=(0, 0), arrowprops=arrow_fmt)  # seta eixo y positiva

# Adiciona legenda e grade
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2 linhas, 2 colunas

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Seno")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosseno")

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("Tangente")

axs[1, 1].plot(x, x**2)
axs[1, 1].set_title("Parábola")

plt.tight_layout()  # ajusta espaçamento
plt.show()
```
